```python
# Real Estate Pricing EDA Project

# Importing Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Settings for visualization
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# 1. Loading the Dataset
df = pd.read_csv("C:/Users/harsh/Downloads/housing_data.csv")
df.columns = df.columns.str.strip().str.lower()  # Clean column names
print("Columns:", df.columns.tolist())
df.head()

# 2. Data Cleaning
print("Missing values before cleaning:\n", df.isnull().sum())
df.drop_duplicates(inplace=True)
df.ffill(inplace=True)
print("Missing values after cleaning:\n", df.isnull().sum())

# 3. Univariate Analysis
sns.histplot(df['saleprice'], kde=True)
plt.title('Distribution of House Prices')
plt.show()

sns.countplot(x='bedroomabvgr', data=df)
plt.title('Number of Bedrooms')
plt.show()

# 4. Bivariate Analysis
sns.scatterplot(x='grlivarea', y='saleprice', data=df)
plt.title('Living Area vs Sale Price')
plt.show()

sns.boxplot(x='bedroomabvgr', y='saleprice', data=df)
plt.title('Bedrooms vs Sale Price')
plt.show()

sns.boxplot(x='fullbath', y='saleprice', data=df)
plt.title('Full Bathrooms vs Sale Price')
plt.show()

# 5. Multivariate Analysis (only numeric columns)
numeric_df = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# 6. Feature Engineering
df['price_per_sqft'] = df['saleprice'] / df['grlivarea']
df['property_age'] = 2025 - df['yearbuilt']

# 7. Feature Engineering and Size Impact
sns.scatterplot(x='grlivarea', y='saleprice', hue='bedroomabvgr', data=df)
plt.title('Living Area vs Sale Price colored by Bedrooms')
plt.show()

sns.boxplot(x='property_age', y='saleprice', data=df)
plt.title('Property Age vs Sale Price')
plt.show()

# 8. Market Trends and Historical Pricing
# If month is in string format (e.g., 'Feb'), convert to numeric first
month_map = {'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'Jun': 6,
             'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12}

if 'yrsold' in df.columns and 'mosold' in df.columns:
    if df['mosold'].dtype == 'object':
        df['mosold'] = df['mosold'].map(month_map)
    df['sale_date'] = pd.to_datetime(dict(year=df['yrsold'], month=df['mosold'], day=1))
    df.set_index('sale_date', inplace=True)
    df['saleprice'].resample('M').mean().plot()
    plt.title('Monthly Average Sale Prices Over Time')
    plt.show()

# 9. Customer Preferences and Amenities
amenity_cols = [col for col in df.columns if col.lower() in ['poolarea', 'garagecars', 'wooddecksf']]
for col in amenity_cols:
    sns.boxplot(x=col, y='saleprice', data=df)
    plt.title(f'{col.capitalize()} vs Sale Price')
    plt.show()

print("EDA complete. Insights ready for reporting!")
```
